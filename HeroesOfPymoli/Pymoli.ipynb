```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "/Users/rohan/Documents/Homework/pandashw/pandas-challenge/02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

```python
# number of players
playerCount = len(df["SN"].unique())
playerCount = pd.DataFrame({"Player Count": [playerCount]})
playerCount
```

```python
# number of unique items
uniqueItems = len(df["Item Name"].unique())
# average price
averagePrice = df["Price"].mean()
# number of purchases
purchases = len(df.index)
# total revenue
revenue = df["Price"].sum()

purchasingAnalysis = pd.DataFrame({"Number of Unique Items": [uniqueItems], "Average Price": [averagePrice], "Number of Purchases": [purchases], "Total Revenue": [revenue]})
purchasingAnalysis["Total Revenue"] = purchasingAnalysis["Total Revenue"].map("$ {:,.2f}".format)
purchasingAnalysis["Average Price"] = purchasingAnalysis["Average Price"].map("$ {:,.2f}".format)
purchasingAnalysis
```

```python
# Gender Count
gender_count_df = df.groupby("Gender")["SN"].nunique()
gender_count_df.head()
```

```python
# Gender Percentage
gender_percentage_df = (gender_count_df/576) * 100
gender_percentage_df.round(2)
```

```python
#Gender demographics DataFrame.
gender_demographics_df = pd.DataFrame({"Gender Count": gender_count_df,"Gender Percentage":gender_percentage_df})
gender_demographics_df["Gender Percentage"] = gender_demographics_df["Gender Percentage"].map("% {:,.2f}".format)
gender_demographics_df
```

```python
#Total number of purchases by gender.


gender_purchases_df = df.groupby("Gender")["Item Name"]
gender_purchases_df.count()
```

```python
# Average price of items by gender.

gender_average_df = df.groupby("Gender")["Price"].mean()
gender_average_df.round(2)
```

```python
#Total Purchase value by gender.

gender_total_df = df.groupby("Gender")["Price"].sum()
gender_total_df
```

```python
#Normalized gender total.
normalized_gender_total_df = gender_total_df/gender_count_df
normalized_gender_total_df.round(2)
```

```python
# Purchasing analysis DataFrame by gender.

gender_analysis_df = pd.DataFrame({"Purchase Count":[113, 652, 15], 
                                   "Average Purchase Price":gender_average_df,
                                   "Total Purchase Value":gender_total_df,
                                   "Normalized Totals":normalized_gender_total_df})
gender_analysis_df
```

```python
#create bins
bins = [0,10,15,20,25,30,35,40, 45]
age_ranges = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
```

```python
# Cut purchase data and place the ages into bins
pd.cut(df["Age"], bins, labels=age_ranges)
```

```python

df["Age Range"] = pd.cut(df["Age"], bins, labels= age_ranges)
df.head()
```

```python
#Players percentage by age range.

age_group_percentage_df = pd.DataFrame(round(df["Age Range"].value_counts()/780,2))
age_group_percentage_df
```

```python
# Purchase count by age range.

age_group_count_df = df.groupby("Age Range")["Item Name"]
age_group_count_df.count()
```

```python
# Average purchase price by age range.

age_group_average_df = df.groupby("Age Range")["Price"].mean()
age_group_average_df.round(2)
```

```python
#Total purchase value by age range.

age_group_total_df = df.groupby("Age Range")["Price"].sum()
age_group_total_df
```

```python
#Normalized totals by age range.

normalized_age_total_df = age_group_total_df/573
normalized_age_total_df.round(2)
```

```python
# Convert to DataFrame.

age_range_df = pd.DataFrame({"Purchase Count":age_group_count_df,
                            "Average Purchase Price":age_group_average_df,
                            "Total Purchase Value": age_group_total_df,
                            "Avg Total Purchase per Person": normalized_age_total_df
})
age_range_df
```

```python
#Overall spending analysis

players_purchase_count_df = df.groupby("SN").count()["Price"].rename("Purchase Count")
players_average_price_df = df.groupby("SN").mean()["Price"].rename("Average Purchase Price")
players_total_df = df.groupby("SN").sum()["Price"].rename("Total Purchase Value")

#Convert to DataFrame.

total_user_data_df = pd.DataFrame({"Purchase Count":players_purchase_count_df,
                                   "Average Purchase Price": players_average_price_df,
                                   "Total Purchase Value": players_total_df})
total_user_data_df.head(10)
```

```python
top_five_spenders = total_user_data_df.sort_values("Total Purchase Value", ascending=False)
top_five_spenders.head()
```

```python
# Total items purchases analysis.

items_purchase_count_df = df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_average_price_df = df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
items_value_total_df = df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

# Convert to DataFrame

items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count_df,
                                   "Item Price":items_average_price_df,
                                   "Total Purchase Value":items_value_total_df,})

items_purchased_df.head()
```

```python
# Sort table to show the five the most popular items.

most_popular_items_df = items_purchased_df.sort_values("Purchase Count", ascending=False)
most_popular_items_df.head()
```

```python
# Sort table to show the five the most profitable items.

most_profitable_items_df = items_purchased_df.sort_values("Total Purchase Value", ascending=False)
most_profitable_items_df.head()
```
